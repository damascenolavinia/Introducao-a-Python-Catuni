```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import unicodedata
```

```python
df= pd.read_csv("Gaming_Hours_vs_Performance_1000_Rows.csv")

df["Stress_Level"] = df["Stress_Level"].astype(int)
df["Daily_Gaming_Hours"] = df["Daily_Gaming_Hours"].astype(int)
print(df["Stress_Level"], df["Daily_Gaming_Hours"])
df.info()
```

### Perguntas
##### 1- O tempo gasto em jogos influencia o nivel de foco dos individuos?

```python
#Pergunta 1
plt.bar(df["Daily_Gaming_Hours"], df["Focus_Level"],color="orange")
plt.title("Horas de jogo vs Nivel de foco")
plt.xlabel("Horas de jogo")
plt.ylabel("Nivel de foco")
plt.show()
```

##### Observando o grafico percebemos que quanto mais as horas de jogo aumentam, mais o nivel de foco diminui. Isso mostra uma correlação entre os fatores, mas nao necessariamente uma relação de causalidade, visto que, nessa situação, o nivel de foco diminuiu a partir das 5 horas de jogo.

### 2- As horas de jogo podem ter relação com os niveis de estresse?

```python
#Pergunta 2
plt.bar(df["Daily_Gaming_Hours"],color="green", height=10)
plt.title("Relacao entre horas de jogo e estresse")
plt.xlabel("Horas de jogo")
plt.ylabel("Niveis de estresse")
plt.show()
```

#### O grafico acima mostra de as horas de jogo influenciam diretamente nos niveis de estresse dos individuos. Isso mostra uma relação de correlação e causalidade, visto que a maioria dos jogos utilizam de violencia e competitividade para alcançar os objetivos.
